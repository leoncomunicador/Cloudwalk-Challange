```python
# Importa o pandas para manipulação dos dados
import pandas as pd
# Importa o matplotlib.pyplot para geração de gráficos
import matplotlib.pyplot as plt
# Importa o seaborn para visualização dos gráficos de forma mais atraente
import seaborn as sns
```

```python
# Importa o sqlite3 para interagir com o banco de dados SQLite
import sqlite3
```

```python
# Lê os arquivos CSV e armazena os dados nos objetos DataFrame df1 e df2

df1 = pd.read_csv('files/checkout_1.csv')
df2 = pd.read_csv('files/checkout_2.csv')
```

```python
# Exibe a mensagem "Dados Checkout 1:"
# Imprime os primeiros registros do DataFrame df1

print("Dados Checkout 1:")
print(df1.head())
```

```python
# Exibe a mensagem "Dados Checkout 2:"
# Imprime os primeiros registros do DataFrame df2

print("Dados Checkout 2:")
print(df2.head())
```

```python
# Exibe a mensagem "Informações estatísticas do Checkout 1:"
# Calcula e imprime informações estatísticas do DataFrame df1

print("Informações estatísticas do Checkout 1:")
print(df1.describe())
```

```python
# Exibe a mensagem "Informações estatísticas do Checkout 2:"
# Calcula e imprime informações estatísticas do DataFrame df2

print("Informações estatísticas do Checkout 2:")
print(df2.describe())
```

```python
# Exibe a mensagem "Correlação do Checkout 1:"
# Calcula a correlação entre as colunas numéricas do DataFrame df1 (exceto a coluna "time")
# Imprime o resultado da correlação

print("Correlação do Checkout 1:")
print(df1.drop('time', axis=1).astype(float).corr())
```

```python
# Exibe a mensagem "Correlação do Checkout 2:"
# Calcula a correlação entre as colunas numéricas do DataFrame df2 (exceto a coluna "time")
# Imprime o resultado da correlação

print("Correlação do Checkout 2:")
print(df2.drop('time', axis=1).astype(float).corr())
```

```python
# Gera um gráfico de linha usando seaborn e matplotlib.pyplot para o DataFrame df1
# O gráfico mostra a variação das vendas ao longo do tempo para as colunas 'today', 'yesterday' e 'same_day_last_week'
# Define título, rótulos dos eixos e legenda do gráfico
# Exibe o gráfico

plt.figure(figsize=(12, 6))
sns.lineplot(data=df1, x='time', y='today', label='Today')
sns.lineplot(data=df1, x='time', y='yesterday', label='Yesterday')
sns.lineplot(data=df1, x='time', y='same_day_last_week', label='Same Day Last Week')
plt.title('Comparação de vendas por hora (Checkout 1)')
plt.xlabel('Hora')
plt.ylabel('Vendas')
plt.legend()
plt.show()
```

```python
# Gera um gráfico de linha usando seaborn e matplotlib.pyplot para o DataFrame df2
# O gráfico mostra a variação das vendas ao longo do tempo para as colunas 'today', 'yesterday' e 'same_day_last_week'
# Define título, rótulos dos eixos e legenda do gráfico
# Exibe o gráfico

plt.figure(figsize=(12, 6))
sns.lineplot(data=df2, x='time', y='today', label='Today')
sns.lineplot(data=df2, x='time', y='yesterday', label='Yesterday')
sns.lineplot(data=df2, x='time', y='same_day_last_week', label='Same Day Last Week')
plt.title('Comparação de vendas por hora (Checkout 2)')
plt.xlabel('Hora')
plt.ylabel('Vendas')
plt.legend()
plt.show()
```

```python
# Cria uma conexão com o banco de dados SQLite e salva os DataFrames df1 e df2 nas tabelas checkout_1 e checkout_2, respectivamente

conn = sqlite3.connect('checkout_data.db')
df1.to_sql('checkout_1', conn)
df2.to_sql('checkout_2', conn)

# Executa uma consulta SQL para obter os dados da tabela checkout_1
# Armazena os dados em um objeto DataFrame chamado checkout_1_data
# Exibe a mensagem "Dados do Checkout 1 (consulta SQL):"
# Imprime os dados do DataFrame checkout_1_data

query = "SELECT * FROM checkout_1"
checkout_1_data = pd.read_sql(query, conn)
print("Dados do Checkout 1 (consulta SQL):")
print(checkout_1_data)

# Executa uma consulta SQL para obter os dados da tabela checkout_2
# Armazena os dados em um objeto DataFrame chamado checkout_2_data
# Exibe a mensagem "Dados do Checkout 2 (consulta SQL):"
# Imprime os dados do DataFrame checkout_2_data

query = "SELECT * FROM checkout_2"
checkout_2_data = pd.read_sql(query, conn)
print("Dados do Checkout 2 (consulta SQL):")
print(checkout_2_data)
```
